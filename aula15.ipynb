Método Apply e ordenação de dataframes

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tested.csv')
df.head()
```

```python
df['Age'].apply(np.sqrt)#aplica uma função a uma coluna específica.
```

```python
def changesurvived(el):
    if el == 0:
        return "No"
    else: 
        return "Yes"

df['Survived'] = df['Survived'].apply(changesurvived)
df.head()
```

```python
def changeFare(el):
    return "${:.2f}".format(el)
   

df['Fare'] = df['Fare'].apply(changeFare)
df.head()
```
