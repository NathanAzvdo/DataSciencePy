```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0,10,0.01)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, x, y2)
plt.grid(True)
```

```python
plt.plot(x,y1, label='sin(x)')
plt.plot(x,y2, label='cos(x)')
plt.legend(loc='upper right')
plt.xticks(range(11))
plt.yticks([-2,0,2])
```

Subplots


```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.plot(x,y1)
plt.subplot(1,3,2)
plt.plot(x,y2)
plt.tight_layout()
```

```python
y3 = np.sin(x) + np.cos(3*x)
y4 = np.sin(2*x)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
```
