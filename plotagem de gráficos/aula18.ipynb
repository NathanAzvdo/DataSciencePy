```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('prs.csv')
df
```

```python
df_py = df[df['name']=='Python']
df_jar = df[df['name']=='Java']

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14,5))
axes[0].plot(df_py['year'], df_py['count'], marker='o')
axes[0].set_xlabel('Ano')
axes[0].set_ylabel('Número Prs')
axes[0].set_title('Número de Prs em python no github')

axes[1].plot(df_jar['year'], df_jar['count'], marker='o')
axes[1].set_xlabel('Ano')
axes[1].set_ylabel('Número Prs')
axes[1].set_title('Número de Prs em Java no github')



```
